### 4. Write a NumPy program to compute the x and y coordinates for points on a sine curve and plot the points using matplotlib.

```python
%matplotlib inline

x = np.arange(0, 3 * np.pi, 0.1) 
y = np.sin(x)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Sine Curve')
plt.plot(x, y)
```
